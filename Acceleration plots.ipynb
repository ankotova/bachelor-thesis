```python
# Visualisation of the Acceleration sensorÂ´s tests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('white')
```

```python
# Data for Figure 11
# Max and Min RSSI

x = open("200_2 (1).txt", "r")
x_2_long = x.readlines()
x_2_long = [int(line.split(".")[1]) for line in x_2_long[:500]] 
x.close()

print(max(x_2_long))
print(min(x_2_long))
```

```python
# Figure 11.1
# plt.style.use('seaborn-whitegrid')
# plt.rcParams['figure.dpi'] = 100 
fig, ax =plt.subplots(1, 1, figsize=(6, 4))
n, bins, patches = plt.hist(x_2_long, facecolor = '#53BEF2', edgecolor='#169acf', linewidth= 1.5, bins = 20)
ax.set_xticks([-90, -88, -86, -84, -82, -80, -78, -76, -74, -72, -70, -68])
# plt.title('RSSI at 2 m distance')
ax.set_ylabel('Amount of telegrams')
ax.set_xlabel('RSSI, dBm')
sns.despine()
fig.savefig('./plot1.png', bbox_inches='tight', dpi=600)
```

```python
# Calculating distance from the received rssi
# Max and Min calculatedd distance
n = 2
rssi0 = -69.86
dist_2_long = 10 ** ((rssi0 - np.asarray(x_2_long))/(10*n))
print(max(dist_2_long))
print(min(dist_2_long))

```

```python
# Figure 11.2
# plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.dpi'] = 100
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.hist(dist_2_long, bins = 20, facecolor = '#53BEF2', edgecolor='#169acf', linewidth= 1.5)
#plt.xticks([1, 2, 3, 4, 5, 6, 11])
# plt.title('Distance calculated out of the RSSI for 2m')
ax.set_ylabel('Amount of telegrams')
ax.set_xlabel('Distance, m')
sns.despine()
plt.savefig('./plot2.png', bbox_inches='tight', dpi=500)
```

```python
# Mean RSSI and calculated distance
n = 2
rssi0 = -69.86
# threshold = 500
mean_rssi = np.mean(x_2_long[:500])
print(mean_rssi)
new_dist = 10 ** ((rssi0 - mean_rssi)/(10*n))
print(new_dist)
```

```python
# Figure 11.3
# RSSI distribution over the time at 2m
time_2_long = np.arange(1, len(x_2_long)+1, 1)
# plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.scatter(time_2_long, x_2_long, s = 50, c = '#53BEF2', alpha = 0.3)

# plt.title("RSSI distribution at 2 m")
ax.set_ylabel("RSSI, dBm")
ax.set_xlabel("Time, min")
sns.despine()
fig.savefig('./plot3.png', bbox_inches='tight', dpi=500)
```
